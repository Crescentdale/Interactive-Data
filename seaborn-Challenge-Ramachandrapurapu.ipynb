# Group 1 Challenge

```python
import pandas as pd 
import seaborn as sns
```

```python
titanic = pd.read_csv('titanic_clean.csv')
```

```python
titanic['Sex'] = titanic['sex'].map({'male': 'Male', 'female' : 'Female'}) 
```

```python
sns.relplot(
    data= titanic, 
    x = 'fare', y = 'age', 
    col ='Sex'
) 

plt.show()
```

```python
sns.relplot(
    data= titanic, 
    x = 'fare', y = 'age', 
    col ='Sex', 
    size='pclass'
) 
plt.show()
```

```python
titanic['Survived_Ind'] = titanic['survived'].map({0: 'Did not survive', 1 : 'Survived'}) 
```

```python
sns.relplot(
    data= titanic, 
    x = 'fare', y = 'age', 
    col ='Survived_Ind', 
    size='pclass',   
    hue='Survived_Ind',
    palette={'Did not survive': '#ff5733', 'Survived': '#33ff57'}
 
) 
plt.show()
```

# Group 2 Challenge 

```python
from matplotlib import pyplot as plt 
```

```python
smh = pd.read_csv('Student Mental health.csv')
```

```python
smh.groupby('Choose your gender')[['Do you have Depression?', 'Do you have Anxiety?',  'Do you have Panic attack?']].count()
```

```python
columns_to_plot = smh.columns[-4:-1] 
category_order = ['Yes','No']
```

```python
for column in columns_to_plot: 
    counts = smh[column].value_counts() 
    counts = counts.reindex(category_order, fill_value=0) 

    print(counts) 
    print()
```

```python
sns.countplot(smh, x='Do you have Depression?', 
              stat='percent', hue='Did you seek any specialist for a treatment?') 
plt.show()
```

```python
sns.countplot(smh, x='Do you have Anxiety?', 
              stat='percent', hue='Did you seek any specialist for a treatment?') 
plt.show()
```

```python
sns.countplot(smh, x='Do you have Panic attack?', 
              stat='percent', hue='Did you seek any specialist for a treatment?') 
plt.show()
```

```python
sns.countplot(smh, x='Do you have Depression?', 
              stat='percent', hue='Choose your gender', palette='PuBuGn') 
plt.show()
```

```python
sns.countplot(smh, x='Do you have Anxiety?', 
              stat='percent', hue='Choose your gender', palette='PiYG') 
plt.show()
```

```python
sns.countplot(smh, x='Do you have Panic attack?', 
              stat='percent', hue='Choose your gender', palette='bwr') 
plt.show()
```
